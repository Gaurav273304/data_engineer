```python
import pandas as pd
```

```python
df = pd.read_csv('employees.csv')
mask = df['salary'] > 50000
filtered_df = df[mask]
# filtered_df.shape

# print(df[df['department'] == 'Engineering'])

# print(df[df['department'] != 'HR'])
print(df[df['age'] > 35])
```

```python
#Filtering multiple columns
# df[(df['department'] == 'Engineering') & (df['salary'] > 100000) & (df['city'] == 'Delhi')]
df[(df['department'] == 'Engineering') & (df['age']< 45)]
```

```python
#filtering with isin
cities = ['Mumbai','Delhi','Bangalore']
# df[df['city'].isin(cities)]
df[~df['city'].isin(cities)] #negation of isin


designations = ['Senior','Intern','VP']
df[df["designation"].isin(designations)]
```

```python
#filteering with between
df[df['age'].between(30,40)]
result = df[df['age'].between(30,40,inclusive="left")]
print(result)
```

```python
#filtering with query
result = df.query('department == "Engineering" and age < 45')
result

```

```python
#what if we want to filter with variables
min_salary = 50000
max_age = 40
result = df.query('salary > @min_salary and age < @max_age')
result
```

```python
#filtering with string methods
result = df[df['name'].str.contains('Sharma')] #case sensitive search
result = df[df['name'].str.contains('Sharma',case=False)] #case insensitive search
result
```

```python
result = df[df["name"].str.startswith('A',case= False)] #case insensitive search for names starting with A
#case sensitive

result
```

```python
result = df[df["name"].str.contains("Kumar",case = False) & (df['department'] == "Engineering")]
result
```
