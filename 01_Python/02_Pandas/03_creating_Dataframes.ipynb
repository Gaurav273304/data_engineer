```python
import pandas as pd
```

```python
# dataframe from a dictionary
df = pd.DataFrame({
    "ID": [1,2,3],
    "Salary": [50000, 60000, 45000],
    "Name" : ["Alice", "Bob", "Charlie"]
},index=["a", "b", "c"])
print(df)
```

```python
#list of dictionaries
df1 = pd.DataFrame(
    [
        {"ID": 1, "Salary": 50000, "Name": "Alice"},
        {"ID": 2, "Salary": 60000, "Name": "Bob"},
        {"ID": 3, "Salary": 45000, "Name": "Charlie"}
    ],index=["a", "b", "c"]
)
print(df1)
```

```python
#Empty dataframe
df2 = pd.DataFrame(columns= ["ID", "Salary", "Name"])
print(df2)
```
