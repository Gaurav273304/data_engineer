## 1) Basics of Indexing in Pandas
DataFrame ke paas row index aur column labels hote hain.
Do tarike hote hain data access karne ke:

##### Label-based → .loc[]
##### Position-based → .iloc[]

```python
import pandas as pd
df = pd.read_csv("employees.csv")
```

```python
print(df.dtypes)
```

```python
print(df["Department"].memory_usage(deep=True))
```

```python
df["Department"] = df["Department"].astype("category")
print(df["Department"].memory_usage(deep=True))
```

```python
df["EmpID"] = df["EmpID"].astype("int32")
df["Salary"] = df["Salary"].astype("float32")
df["JoiningDate"] = pd.to_datetime(df["JoiningDate"])
```

```python
df.dtypes
```

### .loc[] → Label-based Selection

```python
df.head(5)
```

```python
# a) Single row by index label
df.loc[2]
```

```python
# b) Multiples row
df.loc[[1,3,5]]
```

```python
# Specific row & column
```

```python
df.loc[2, "Name"]
```

```python
#Multiple rows & columns
```

```python
df.loc[[1, 3], ["Name", "Salary"]]
```

```python
#Conditional selection with loc
df.loc[df["Department"] == "IT",["Name","Salary"]]
```

### .iloc[] → Position-based Selection

```python
#Single row by position
```

```python
df.iloc[0]
```

```python
df.iloc[[0, 2, 3]]
```

```python
# Row & Column both
```

```python
df.iloc[1, 2]
```

```python
df.head(4)
```

```python
#Range slicing
```

```python
df.iloc[0:3,1:3]
```

### Slicing with Index

```python
# Row slicing (works on labels if index sorted)
```

```python
df.loc[1:2]
```

```python
#Column slicing
```

```python
df.loc[:, "Name":"Salary"].head(5)
```

```python
# Mixed slicing (rows + cols)     Edge Cases & Pitfalls
                                   #.loc includes end index in slicing
df.loc[2:4, "Name":"Department"]
```

```python
df.head(5)
```

```python
df.loc[(df["Department"] == "HR"), ["Name", "Salary"]]
```
