Day 2: Data types (astype()), conversions (pd.to_datetime())

CustomerID ko string me convert karo.

Sales ko float me convert karo aur verify karo dtype.

PurchaseDate ko datetime me convert karke sirf year aur month nikal ke alag column banao.

Check karo Age column ka dtype aur use string me convert karke pehle 5 rows print karo.

```python
df.info()
```

```python
df["CustomerID"] = df["CustomerID"].astype("float")
```

```python
df["Sales"] = df["Sales"].astype("float")
```

```python
df.info()
```

```python
df["PurchaseDate"] = pd.to_datetime(df["PurchaseDate"])
```

```python
df.info()
```

```python
df["Purchase_Month"] = df["PurchaseDate"].dt.month
df["Purchase_Year"] = df["PurchaseDate"].dt.year
```

```python
df.head(5)
```

```python
df["Age"] = df["Age"].astype("float")
```

Day 4: Boolean indexing & multi-condition filters

Un customers ko filter karo jinke Age > 30 hai.

Sirf un customers ko select karo jo City = "Delhi" me hain.

Customers filter karo jinke City = "Delhi" aur Age > 30 hain.

Customers nikal ke dikhao jinke Sales > 8000 ya phir Product = "Laptop" hai.

```python
df.info()
```

```python
df["Age"]  = df["Age"].astype("Int64")
```

```python
over_thirty = df[df["Age"] > 30].head(4)
over_thirty
```

```python
delhi = df[df["City"]=="Delhi"]
delhi.head(4)
```

```python
df.query("Age > 30 and City == 'Delhi'").head(5)
```

```python
#Customers nikal ke dikhao jinke Sales > 8000 ya phir Product = "Laptop" hai.
df.query("Sales > 8000 and Product == 'Laptop'").head(5)
```
