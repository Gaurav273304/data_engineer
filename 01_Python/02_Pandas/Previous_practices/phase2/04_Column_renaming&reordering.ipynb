```python
import pandas as pd

data = {
    "Customer ID": [1, 2, 3],
    "Full Name": ["Ravi Kumar", "Sita Devi", "Aman Verma"],
    "TotalSales(₹)": [2000, 3000, 2500],
    "Join Date": ["2023-01-01", "2023-02-15", "2023-03-20"]
}

df = pd.DataFrame(data)
print("Original DF:\n", df)
```

# Column Renaming
### a) Rename single column

```python
df.rename(columns={'Customer ID': 'customer_id'},inplace=True)
```

```python
df.rename(columns={"Full Name": "full_name","TotalSales(₹)": "total_sales","Join Date": "join_date"},inplace=True)
df
```

### c) Rename using str methods (shortcuts)

```python
# All column names to lowercase
df.columns = df.columns.str.lower()
# Replace spaces with underscore
df.columns = df.columns.str.replace(" ", "_")
# Remove special characters (like ₹, (), etc.)
df.columns = df.columns.str.replace(r'[^a-zA-Z0-9_]', '', regex=True)

```

# Function: Convert to Snake Case

```python
import re
import pandas as pd

def to_snake_case(df):
    """
    Convert all dataframe column names to snake_case
    and remove special characters.
    """
    new_cols = []
    for col in df.columns:
        # Lowercase
        col = col.lower()
        # Replace spaces and hyphens with underscore
        col = re.sub(r'[\s\-]+', '_', col)
        # Remove special characters except underscore
        col = re.sub(r'[^a-z0-9_]', '', col)
        # Remove leading/trailing underscores
        col = col.strip('_')
        new_cols.append(col)
    df.columns = new_cols
    return df
```

```python
data = {
    "Customer ID": [1, 2, 3],
    "Full Name": ["Ravi Kumar", "Sita Devi", "Aman Verma"],
    "TotalSales(₹)": [2000, 3000, 2500],
    "Join-Date": ["2023-01-01", "2023-02-15", "2023-03-20"]
}
```

```python
df = pd.DataFrame(data)
print("Original columns:", df.columns.tolist())
```

```python
df = to_snake_case(df)
print("Snake_case columns:", df.columns.tolist())
```
