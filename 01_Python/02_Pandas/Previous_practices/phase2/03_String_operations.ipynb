```python
import pandas as pd

data = {
    "customer_id": [1, 2, 3, 4, 5, 6],
    "name": ["Ravi Kumar", "Sita Devi", "Aman Singh", "Neha", "Rohit123", "Pooja Sharma"],
    "email": ["ravi@gmail.com", "sita@yahoo.com", "aman@company.com", 
              "neha_gmail.com", "rohit@outlook.com", "pooja@@gmail.com"],
    "phone": ["+91-9876543210", "98765 43211", "123-456-7890", "99999-88888", "0000000000", "91-9898989898"],
    "address": ["Delhi, India", "Mumbai", "Noida, UP", "Bangalore, KA", "Pune", "Hyderabad, TS"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.info()
```

# 1. .str.contains()

#### Example 1: Emails me "gmail" hai?

```python
df["is_gmail"] = df["email"].str.contains("gmail",na = False)
df["is_gmail"]
```

### Example 2: Phone number India ka (+91) hai?

```python
df["is_ind_num"] = df["phone"].str.contains(r'^\+91',regex = True)
df["is_ind_num"].sum()
```

# 2. .str.replace()
Strings me unwanted characters ya patterns ko replace karna.

```python
#Example 1: Phone numbers clean karna (digits only)
df['clean_phone'] = df['phone'].str.replace(r'\D', '', regex=True)
df
```

```python
#Example 2: Names me numbers hatao
df["clean_name"] = df["name"].str.replace(r'\d+','',regex = True)

df
```

# 3 Regex with .str.extract()

```python
#Example 1: Email se domain nikalna
df['domain'] =  df['email'].str.extract(r'@(\w+\.\w+)')
df
```

#### Lowercase / Uppercase

```python
df['name_lower'] = df['name'].str.lower()
df['name_upper'] = df['name'].str.upper()
df
```

#### Substring split (e.g., first name, last name)

```python
df[['first_name', 'last_name']] = df['name'].str.split(" ", n=1, expand=True)# expand = True ka matlab hai ki result alag alag column me ayenge,n=1 matlab ek hi split hoga wo bhi pahle space ke bad
df
```

```python
df['valid_email'] = df['email'].str.contains(r'@.+\..+', regex=True)
df
```

```python
df['provider'] = df["email"].str.extract(r'@(\w+\.\w+)')
df
```

```python
#Phone numbers ko standardize karna (sirf 10 digits chhodo).
df["clean_phone"] = df['phone'].str.replace(r'\D', '', regex=True)
df["standard_phone"] = df["clean_phone"].str[-10:]
df
```

```python
df = df.drop(columns = ["provider"])
df
```
