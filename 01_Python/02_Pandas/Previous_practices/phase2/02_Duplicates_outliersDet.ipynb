```python
import pandas as pd

data = {
    "customer_id": [1, 2, 3, 4, 5, 2, 6, 7, 8, 9, 10],
    "name": ["Ravi", "Sita", "Aman", "Neha", "Rohit", "Sita", "Anil", "Pooja", "Asha", "Karan", "Meena"],
    "email": ["ravi@x.com", "sita@x.com", "aman@x.com", "neha@x.com", "rohit@x.com",
              "sita@x.com", "anil@x.com", "pooja@x.com", "asha@x.com", "karan@x.com", "meena@x.com"],
    "salary": [50000, 60000, 55000, 62000, 70000, 60000, 58000, 65000, 59000, 61000, 1000000], # outlier salary
    "sales": [200, 300, 250, 400, 350, 300, 280, 500, 290, 310, 5000]  # outlier sales
}

df = pd.DataFrame(data)
print(df)
```

# Part 2: Outlier Detection

## Method 1: IQR (Interquartile Range)

```python
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df['sales'] < lower_limit) | (df['sales'] > upper_limit)]
outliers
```

```python
df
```

## Method 2: Z-Score (Standard Deviation method)

```python
import numpy as np

mean_sales = df['sales'].mean()
std_sales = df['sales'].std()
df['z_score_sales'] = (df['sales'] - mean_sales) / std_sales
outliers_z = df[df['z_score_sales'].abs() > 3]
print(outliers_z)
```

# Part 3: Cleaning Outliers

### Remove Outliers

```python
df_clean = df[(df['sales'] >= lower_limit) & (df['sales'] <= upper_limit)]
df_clean
```

# real-world ETL style example

```python
import pandas as pd

data = {
    "customer_id": [1, 2, 3, 4, 5, 6],
    "name": ["Ravi", "Sita", "Sita", "Neha", "Ravi", "Aman"],
    "email": ["ravi@x.com", "sita@x.com", "sita@x.com", "neha@x.com", "ravi@x.com", "aman@x.com"],
    "sales": [200, 300, 300, 400, 200, 250]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
df_clean = df.groupby("email", as_index=False).agg({
    "name": "first",        # pehla naam rakha (ya mode bhi le sakte ho)
    "sales": "sum"          # sales ko aggregate (sum) kar diya
})
print("\nDeduplicated & Consolidated Table:\n", df_clean)

# Duplicate emails hat gaye.

# Har email ke liye ek hi row rakh li gayi.

# Sales column ko aggregate (sum) karke consolidate kar diya.

# Ravi ke 2 transactions (200 + 200 = 400)

# Sita ke 2 transactions (300 + 300 = 600)
```
