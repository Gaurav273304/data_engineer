```python
import pandas as pd
```

```python
#Accessing elements in a series using index labels
s = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])
print(s[["b","d"]])
```

```python
#Access by position
print(s.iloc[1])
print(s.iloc[[1,3]])
print(s.iloc[1:3])
```

```python
#operations on series
s1 = pd.Series([1,2,3,4,5])
print(s1 + 10)
print(s1*2)
print(s1[s1>3])
```

```python
#attributes
print(s1.values)
print(s1.index)
print(s1.dtype)
```
