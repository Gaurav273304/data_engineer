# Missing Values Detect karna – isna() / isnull()

```python
df.isnull().sum()
```

## Missing Values Fill karna – fillna()

Common approaches: mean, median, mode, forward-fill, backward-fill.

```python
median_sal = df['salary'].median()
df['salary'] = df['salary'].fillna(median_sal)
df.head(15)
```

# Missing Values Drop karna – dropna()

```python
df.dropna(inplace=True)
df.head(10)
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ravi", "Sita", "Aman", "Neha", "Rohit"],
    "Age": [25, np.nan, 30, 28, np.nan],
    "Salary": [50000, np.nan, 60000, np.nan, 75000],
    "Dept": ["IT", "HR", "Finance", "IT", "Finance"]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df['Salary'].dtype)
```

```python
print(df['Salary'].median())
```

```python
mean_sal = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_sal).round()
df
```
