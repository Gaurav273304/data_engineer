```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('messy_customers.csv')
df.info()
```

## Load & Initial Exploration

```python
df.describe()
```

```python
df.head(5)
```

## Check Column Names & Rename

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
df.head(4)
```

```python
df['joindate'] = pd.to_datetime(df['joindate'],errors = 'coerce')
df.info()
```

```python
df[['city','name']] = df[['city','name']].astype(str)
df.info()
```

# Missing Data Handling

```python
df.isna().sum()
```

```python
df['joindate'] = df['joindate'].ffill()
df['age'] = df['age'].ffill()
df.isna().sum()
```

```python
df['joindate'] = df['joindate'].bfill()
df.isna().sum()
```

# Duplicate Records Check

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

# Outlier Detection

```python
Q1 = df['total_sales'].quantile(0.25)
Q3 = df['total_sales'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['total_sales'] >= Q1-1.5*IQR) & (df['total_sales'] <= Q3+1.5*IQR)]
df_clean = df[(df['total_sales'] >= Q1 - 1.5 * IQR) & (df['total_sales'] <= Q3+1.5*IQR) ]
df_clean.info()
```

# String Cleaning & Regex

```python
df['email'].sample(20)
```

```python
df['email_domain'] = df['email'].str.extract(r'@(\w+\.\w+)')
df.head(10)
```

```python
df['join_month'] = df['joindate'].dt.month_name()
df.head(10)
```

# Filtering & Boolean Indexing

```python
df[(df['age'] > 30) & (df['city'] == 'Delhi') & (df['join_month'] == 'January')]
```
