## .astype() – Convert Data Types

### Syntax df["column"] = df["column"].astype(new_dtype)

```python
data = {"ID": ["1", "2", "3"], "Salary": ["50000", "60000", "45000"]}
df = pd.DataFrame(data)
print(df.dtypes)
```

```python
df["ID"] = df["ID"].astype(int)
df["Salary"] = df["Salary"].astype(float)
print(df.dtypes)
```

### pd.to_datetime() – Date Conversion

```python
dates = pd.Series(["2023-01-10", "2023/02/15", "March 5, 2023"])
dt = pd.to_datetime(dates,format = 'mixed')
print(dt)
```

### CSV Load + Type Conversion (Exercise Style)

```python
df = pd.read_csv("employees.csv")
print(df.dtypes)
```

```python
df["EmpID"] = df["EmpID"].astype("Int64")
df["Salary"] = df["Salary"].astype("float32")
df["JoiningDate"] = pd.to_datetime(df["JoiningDate"])
df["Department"] = df["Department"].astype("category")
```

```python
print(df.dtypes)
```

```python
df["joining_Year"] = df["JoiningDate"].dt.year
df["joining_Month"] = df["JoiningDate"].dt.month
df
```

```python
average_salary = df["Salary"].mean()
average_salary
```
