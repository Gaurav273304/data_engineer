# Sorting in Pandas

### sort_values() → Sort by column values

```python
df.sort_values(by = "Salary")  #ascending order
```

```python
df.sort_values(by = "Salary",ascending = False)
```

```python
#Multiple columns(Tie breaker)
df.sort_values(by=["Department", "Salary"], ascending=[True, False])
```

### sort_index() → Sort by row/column index

```python
df.sort_index()  
```

```python
df.sort_index(axis=1)   # Sort columns alphabetically
```

```python
top5 = df.nlargest(5,"Salary")
top5
```

```python
bottom3 = df.nsmallest(3,"Salary")
bottom3
```

```python
df1 = df.sample(frac = 0.2)
df1
```

```python
df.sort_index(axis=1)
```

```python
top3= df.nlargest(3,"Salary")
top3
```

# Sampling in Pandas

#### Random sample of rows

```python
df.sample(n=3)
```

#### Fractional sampling

```python
df.sample(frac=0.2)
```

```python
# With replacement
df.sample(n=5, replace=True)
```

```python
# Random state (reproducibility)
df.sample(n=3, random_state=42)
```

#### Column-wise sample

```python
df.sample(n=2, axis=1,random_state=42 )   # 2 random columns
```

# Practical Example DataFrame

```python
import pandas as pd

data = {
    "OrderID": range(1, 11),
    "Customer": ["Ravi","Sita","Amit","Neha","Kiran","Pooja","Vikas","Rani","Arjun","Meena"],
    "Sales": [5000, 7000, 12000, 8000, 15000, 2000, 9500, 11000, 4000, 18000]
}

df = pd.DataFrame(data)
print(df)
```

```python
top10 = df.sort_values(by="Sales",ascending = False).head(10)
top10
```

```python
df.nlargest(5, "Sales")
```

```python
df.nsmallest(5, "Sales")
```
