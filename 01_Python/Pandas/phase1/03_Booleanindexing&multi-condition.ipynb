```python
import pandas as pd
import numpy as np

# Sample data
data = {
    "EmpID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Name": ["Ravi", "Sita", "Amit", "Neha", "Arjun", "Meera", "Karan", "Pooja"],
    "JoiningDate": [
        "2021-05-10", "2022-07-15", "2021-08-01", "2023-01-20",
        "2020-11-11", "2019-03-25", "2021-12-05", "2022-09-30"
    ],
    "Salary": [50000, 60000, 55000, 48000, 75000, 42000, 58000, 67000],
    "Department": ["IT", "HR", "Finance", "IT", "Finance", "HR", "IT", "Finance"],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Adding Age column with random values between 22 and 60
df["Age"] = np.random.randint(22, 60, size=len(df))

# Adding more rows
new_data = {
    "EmpID": [9, 10],
    "Name": ["Raj", "Simi"],
    "JoiningDate": ["2022-10-20", "2021-04-11"],
    "Salary": [53000, 59000],
    "Department": ["IT", "HR"],
    "Age": [32, 45],
}

# Convert new data to DataFrame and append to existing data
new_df = pd.DataFrame(new_data)
df = pd.concat([df, new_df], ignore_index=True)

# Save to CSV
df.to_csv("employees_new.csv", index=False)

print("CSV file created successfully!")
```

## Boolean Indexing Basics
##### Boolean indexing matlab ek True/False mask create karke uske basis pe rows filter karna.

```python
df["Age"] > 30
```

### Single Condition Filtering

```python
df[df["Age"] > 30]
```

```python
df[df["Salary"]>55000]
```

##### Multiple Conditions
###### Important: Conditions ko ()` bracket ke andar likhna zaruri hai, warna error aayega.

```python
df[(df["Age"] > 25) & (df["Department"] == "HR")]
```

```python
df[(df["Department"] == "HR") | (df["Department"] == "IT")]
```

```python
df[~(df["Department"] == "Finance")]
```

```python
import pandas as pd

data = {
    "CustomerID": [1, 2, 3, 4, 5, 6],
    "Name": ["Ravi", "Sita", "Amit", "Neha", "Kiran", "Pooja"],
    "Age": [25, 40, 32, 29, 35, 42],
    "City": ["Delhi", "Mumbai", "Delhi", "Delhi", "Bangalore", "Delhi"]
}

df = pd.DataFrame(data)
print(df)
```

#### isin() → multiple values filter

```python
df[df["City"].isin(["Delhi","Mumbai"])]
```

### Large datasets ke liye multiple filters chain karne ki jagah ek hi .query() use kar sakte ho:

```python
df.query("Age > 30 and City == 'Delhi'")
```

```python
result = df[(df["Age"] > 30) & (df["City"] == "Delhi")]
print(result)
```

```python
df.query("Age < 35 and City == 'Delhi'")
```
