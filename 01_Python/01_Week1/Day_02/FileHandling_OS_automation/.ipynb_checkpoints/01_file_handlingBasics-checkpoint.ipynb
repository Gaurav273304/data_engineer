```python
#count how many charecters are there
f = open("hello.txt", "r")
'''
# method 1
count = 0
for line in f:
    count += len(line)
print(count)
'''
r = f.read()
count = 0
for i in r:
    if i!= '\n' and i!= " ":
        count += 1
print(count)
f.close()
```

```python
#reverse the each line of the file
f = open("hello.txt",'r')
r = f.readlines()
for i in r:
    s = i.split()
    print(" ".join(s[::-1]))
f.close()
```

```python
#return the number of words present in the file

def count_words(file):
    file_open = open(file,'r')
    file_read = file_open.read()
    return len(file_read.split())

print(count_words('asess.txt'))
```

```python
# return the number of lines in a file
def count_lines(filename):
    file_open = open(filename,'r')
    file_read = file_open.read()
    return len(file_read.split('\n'))

print(count_lines('asess.txt'))
    
```

```python
def count_words(file):
    file_open = open(file,'r')
    file_read = file_open.read()
    x = file_read.split()
    for i in x:
        if len(i) > 4:
            print(i,end = " ")
count_words('asess.txt')
```

```python
#reverse the each word in a string
f = open('asess.txt','r')
r = f.readlines()
for i in r:
    rev = i.split()
    temp = []
    for i in rev:
        temp.append(i[::-1])
    print(" ".join(temp))
```
