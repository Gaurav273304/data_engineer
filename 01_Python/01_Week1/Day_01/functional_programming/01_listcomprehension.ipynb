# List Comprehension

### Syntax -> [expression for item in iterable if condition]
expression → kya banana hai

item → ek-ek element of iterable

iterable → list, tuple, dict, pandas column, etc.

if condition → optional (filter karne ke liye)

```python
nums = [1, 2, 3, 4]
squares = [x*x for x in nums]
print(squares) 
```

```python
nums = [1, 2, 3, 4, 5]
evens = [x for x in nums if x % 2 == 0]
print(evens)   # [2, 4]
```

```python
pairs = [(x,y) for x in [1,2] for y in ['a','b']]
print(pairs)
```

```python
data = ["apple", "", "banana", None, "grape"]
cleaned = [i for i in data if i]  
print(cleaned)  # ['apple', 'banana', 'grape']
```

### Pandas me List Comprehension

Pandas me mostly tum columns ko transform karne ke liye ya naye column banane ke liye
list comprehension use karte ho.

```python
import pandas as pd
df = pd.DataFrame({
    'Name': ['Aparichit', 'Rohit', 'Sita'],
    'Age': [25, 30, 28]
})

df['AgeCat'] = ['Young' if age < 26 else 'Adult' for age in df['Age']]
print(df)
```

```python
df['AgeCat'] = df['Age'].apply(lambda x:'Young' if x<=28 else 'Adult')
df
```

#### Multiple columns combine

```python
df['Info'] = [f"{n}-{a}" for n ,a in zip(df['Name'],df["Age"])]
print(df)
```

```python
# List of only adults
adults = [name for name, age in zip(df['Name'], df['Age']) if age >= 28]
print(adults)  # ['Rohit', 'Sita']
```

```python
def age_bracket(age):
    if age < 26:
        return "Junior"
    elif age < 30:
        return "Mid"
    else:
        return "Senior"

df['Bracket'] = [age_bracket(a) for a in df['Age']]
print(df)
```

```python
df['post'] = ['young_1' if age <26 else 'Adult_1' if age < 30 else 'Senior_1' for age in df['Age']]
df
```
