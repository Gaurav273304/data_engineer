#### syntax -> filter(function, iterable)

```python
def is_even(x):
    return x %2 == 0

nums = [1,2,3,4,5,6]
result11 = filter(is_even,nums)
print(list(result11))
```

```python
result = list(filter(lambda x:x%2==0,nums))
result
```

```python
#  1. Positive numbers filter karna
nums = [-1,-3,0,3,4]
positives = list(filter(lambda x : x>0,nums))
positives
```

```python
#Ye DataFrame ke rows ya columns ko labels ke basis par filter karta hai, values ke basis par nahi.
```

```python
import pandas as pd
df = pd.DataFrame({
    'Name': ['Aparichit', 'Rohit', 'Sita'],
    'Age': [25, 30, 28],
    'City': ['Delhi', 'Mumbai', 'Pune']
})
print(df.filter(items=['Name','Age']))
```

```python
print(df.filter(regex='^C'))
```

```python
print(df.query("Age < 28"))
```
