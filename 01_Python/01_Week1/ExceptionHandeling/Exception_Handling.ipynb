```python
try:
    x = 10 / 0
except:
    print("Error: Division by zero")
```

```python
#Specific Exception Handling
try:
    num = int("abc")   # ValueError
except ValueError:
    print("Invalid number format")
```

```python
#Multiple Exceptions
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid value")
```

```python
#finally Block
try:
    f = open("test.txt", "r")
    data = f.read()
except FileNotFoundError:
    print("File not found")
finally:
    print("Closing operation done")
```

```python
#Raising Custom Exception
def check_age(age):
    if age < 18:
        raise ValueError("Age must be at least 18")
    return "Access granted"

try:
    print(check_age(66))
except ValueError as e:
    print("Error:", e)
```

```python
# division Calculator
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Division by zero not allowed"
    except TypeError:
        return "Only numbers are allowed"
    finally:
        print("Operation attempted")

print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # Division by zero not allowed
print(safe_divide(10, "x")) # Only numbers are allowed
```

### must-know exception handling techniques

# 1. File Handling Exceptions

```python
try:
    with open("data.txt", "r") as f:
        data = f.read()
except FileNotFoundError:
    print("File not found, please check path")
except PermissionError:
    print("No permission to read the file")
# Real life: CSV/JSON config files, log files, input-output data pipelines.
```

# 2. ZeroDivisionError

```python
try:
    result = 100 / 0
except ZeroDivisionError:
    result = None
    print("Division by zero not allowed")
```

# 3. ValueError & TypeError

```python
try:
    age = int("abc")
except ValueError:
    print("Invalid age input, must be number")
```
