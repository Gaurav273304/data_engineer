##### Q1. Remove duplicates from list (without using set)

```python
list1 = [1,2,3,4,5,2,4]
list2 = []
for i in list1:
    if i not in list2:
        list2.append(i)
print(list2)
```

```python
unique_comp = [list1[i] for i in range(len(list1)) if list1[i] not in list1[:i]]
print(unique_comp)
```

##### Q2Generate dictionary of word counts (from a sentence)

```python
from collections import Counter
sentence = "data engineering is fun and data is powerful"
words_count = Counter(sentence.split())
print(words_count)
```

##### Q3.Find numbers divisible by both 3 and 5 (1–100)

```python
numbers = [i for i in range(1,100) if i%3 == 0 and i%5 == 0]
print(numbers)
```

#### Q4.Create list of squares using map()

```python
#normal method
num = [1,2,3,4,5]
squred = [i**2 for i in num]
print(squred)
```

```python
#using map function
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

##### Q4.Filter names starting with “A”

```python
names = ["Aparichit", "Ravi", "Anjali", "Rohan", "Aditi"]
filtered = list(filter(lambda name: name.startswith('A'), names))
print(filtered)
```

```python
#using loop

def filterd(names):
    for name in names:
        if name.startswith("A"):
            return name
filterd_namess =  list(filter(filterd,names))
print(filterd_namess)
```

##### Q5.Flatten a nested list

```python
nested = [[1,2], [3,4,5], [6]]
flattened = [num for sublist in nested for num in sublist]
print(flattened)
```

```python
a = [[1,2], [3,4,5], [6]]
b = []
for i in a:
    for j in i:
        b.append(j)
print(b)

```
