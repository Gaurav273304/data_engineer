## map

```python
# functions ke sath
def square(x):
    return x*x
nums = [1,2,3,4]
result = map(square,nums)
print(list(result))
```

```python
# lambda ke sath

nums1 = [11,22,33,44]
res = list(map(lambda x:x+4,nums1))
res
```

```python
# multiple iterables
a = [1, 2, 3]
b = [4, 5, 6]

result = list(map(lambda x,y:x+y,a,b))
print(result)
```

```python
#Strings ko uppercase karna
words = ["python","map","function"]
result = list(map(str.upper,words))
result
```

```python
# Pandas me map()
import pandas as pd
data = pd.Series([1,2,3,4])
res = data.map(lambda x:x**2)
res
```

```python
#Custom function ke saath
def label_age(x):
    return "Young" if x < 30 else "Adult"

ages = pd.Series([25,30,40,22])
result = ages.map(label_age)
result
```
