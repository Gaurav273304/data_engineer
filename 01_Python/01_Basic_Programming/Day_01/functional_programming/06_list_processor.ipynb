```python
# Take a list of numbers
# Remove duplicates
# Square each unique number
# Return only even squares

list1 = [2,3,4,5,6,7,8]
unique_list = []
for x in list1:
    if x not in unique_list:
        unique_list.append(x)

square = [i**2 for i in unique_list]
even_square = [i for i in square if i%2==0]
      
print(unique_list)
print(square)
print(even_square)
```
