```python
def write_into_file(userinput,filename):
    with open(filename+".txt","w") as file:
        file.write(userinput)

userInput=input("Enter your file content")
filename=input("Enter your file name")
write_into_file(userInput,filename)
```

```python
# multiple line store
def write_into_file1(userinput,filename):
    with open(filename+".txt","w") as file:
        file.write(userinput)
file_name = input("Enter filename without extension")
user_inp = ""
while True:
    user_i = input("Enter Data = ")
    if user_i == 'q' or user_i == 'Q':
        break
    user_inp += user_i
    user_inp += "\n"

write_into_file1(user_inp[:-1],file_name)
```
