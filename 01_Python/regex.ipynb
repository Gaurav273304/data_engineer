```python
import re

# Basic Methods
# re.search(pattern, string)   # Pattern ko search karta hai
# re.match(pattern, string)    # String ke start se match karta hai
# re.findall(pattern, string)  # Sab matches ki list return karta hai
# re.sub(pattern, repl, string)# Pattern ko replace karta hai
```

```python
text = "I love Python"
pattern = "Python"

result = re.search(pattern, text)
print(result)
```

```python
#2. Find All Matches
text = "cat mat bat rat"
matches = re.findall(r"[m]at", text)
print(matches)
```

```python
text = "hello world"
print(re.findall(r"^hello", text)) # Start
print(re.findall(r"world$", text)) # End
```

```python
text = "My number is 9876543210"
numbers = re.findall(r"\d+", text)
print(numbers)
```

```python
import pandas as pd

data = {
    "email": ["test@example.com", "user@yahoo.in", "coder@outlook.com", "bademail@wrong", "admin@gmail.com"],
    "phone": ["+91-9876543210", "91 8765432109", "(+91) 9998887776", "12345", "0987654321"],
    "name": ["John Doe", "Jane Smith", "Mike O'Neil", "Anna-Marie", "Tom"]
}
df = pd.DataFrame(data)
print(df)


```

## Extract Email Domain (Provider)

Goal: Email ke @ ke baad ka domain nikalna.
Sample: test@example.com → example.com

```python
df["domain"] = df["email"].str.extract(r'@([\w]+\.\w+)')
print(df[["email", "domain"]])
```

```python
df['clean'] = df['phone'].str.replace(r'\D','',regex = True)
df
```
